```python
import PyPDF2
import pytesseract
import shutil
import os
import random
import cv2
from google.colab.patches import cv2_imshow
from IPython.display import Image
```

```python
filename=r'/content/drive/MyDrive/Colab Notebooks/hotsite (1).pdf'
pdfFileObj = open(filename,'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
num_pages = pdfReader.numPages
info=pdfReader.getDocumentInfo()
```

```python
print("No. of Pages: ", num_pages)
print("Titel: ", info.title)
print("Author: ",info.author)
print("Subject: ",info.subject)
print("Creator: ",info.creator)
print("Producer: ",info.producer)
```

```python
count = 0
text = ""
#The while loop will read each page.
while count < num_pages:
 pageObj = pdfReader.getPage(count)
 count +=1
 text += pageObj.extractText()
 print("Page Number",count)
 print("Content",text)
pdfFileObj.close()
```

```python
image_Filename=r'/content/drive/MyDrive/Colab Notebooks/etiqueta.jpg'
```

```python
# Read the file using opencv and show the image
img=cv2.imread(image_Filename)
cv2_imshow(img)
cv2.waitKey(0)
```

```python
#set the configuration for redaing text from image using pytesseract
extractedInformation = pytesseract.image_to_string(img)
print(extractedInformation)
```
